# Exploratory Analysis – ENADE 2022 (UFJF)

This notebook performs an exploratory analysis of ENADE 2022 data for UFJF
courses. The dataset is derived from the same data source used in the
Ontology_ENADE project and is analyzed here from a predictive perspective.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/raw/cursos_ufjf_enade2022.csv")
df.head()
```

```python
df.shape
df.info()
df.columns
```

```python
df.isna().sum().sort_values(ascending=False)
```

```python
sns.histplot(df["NT_GER"].dropna(), kde=True)
plt.title("Distribution of ENADE General Scores (UFJF 2022)")
plt.xlabel("NT_GER")
plt.ylabel("Frequency")
plt.show()
```

```python
df.groupby("CO_CURSO")["NT_GER"].mean().sort_values(ascending=False)
```

## Initial Observations

- The dataset contains X courses from UFJF in ENADE 2022.
- There is variability in ENADE scores across courses.
- Some variables present missing values and will require preprocessing.
- The distribution of scores suggests that both classification and regression
  approaches are feasible.
